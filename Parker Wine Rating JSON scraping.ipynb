```python
import pandas as pd
from sqlalchemy import create_engine

```

```python
import requests
import json

```

```python
url = "https://public.opendatasoft.com/api/records/1.0/search/?dataset=parker&rows=1575&sort=rating&facet=region&facet=year&facet=value_class&facet=maturity"
```

```python
response = requests.get(url).json()
print(json.dumps(response, indent=4, sort_keys=True))


```

```python
wine_list = response["records"]
wine_list


```

```python
wine_list[0]['fields']['rating']
```

```python
# set up lists to hold response data for country, rating, maturity, region, title, year. 
country = []
rating = []
maturity = []
region = []
title = []
year = []

# loop through json
for x in wine_list:
    if 'maturity' in x['fields']:
        maturity.append(x['fields']['maturity'])
    else:
        maturity.append(None)
    if 'rating' in x['fields']:
        rating.append(x["fields"]['rating'])    
    else:
        rating.append(0)
    country.append(x["fields"]['country'])
    region.append(x["fields"]['region'])
    title.append(x["fields"]['value_title'])
    year.append(x["fields"]['year'])
    
# print(maturity)
    
```

```python
len(rating)
```

```python
len(title)
```

```python
lists = {"Country": country, "Rating": rating, "Maturity": maturity, "Region": region, "Title": title, "Year": year}
```

```python
wine_df = pd.DataFrame(lists)
wine_df.head()
```

```python
grouped_by_country = wine_df.groupby("Country")
grouped_by_country.count().head()
```

```python
score_averages = grouped_by_country["Rating"].mean()
score_averages
```

```python
score_averages.to_csv("Wine_rating.csv", index=True, header=True)
```
